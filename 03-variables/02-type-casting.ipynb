## Introduction to Type Casting in Python

Type casting in Python means changing a value from one type to another, like turning a string into a number. You can use functions like `int()`, `float()`, `str()`, and `bool()` to do this. Type casting helps your code work correctly when you need different types.

```python
# Converting user input (string) to integer for calculation
age_str = input('Enter your age: ')  # type: str
print(f"Type of age_str: {type(age_str).__name__}")  # Confirming the type is str

age = int(age_str)  # type casting from str to int
print(f"Type of age: {type(age).__name__}")  # Confirming the type is now int

years_until_30 = 30 - age
print(f"You will be 30 in {years_until_30} years (current age: {age}).")
```

```python
# Example: Converting a float to an integer (e.g., price rounding)
price_str = "19.99"
price_float = float(price_str)  # type casting from str to float
price_int = int(price_float)    # type casting from float to int (removes decimals)
print(f"Rounded down price: {price_int}")

# Example: Converting a number to a string (e.g., for display)
score = 85
score_str = str(score)  # type casting from int to str
print("Your score is: " + score_str)

# Example: Converting a value to boolean (e.g., checking if input is empty)
user_input = ""
is_input_provided = bool(user_input)  # type casting from str to bool
print(f"Did the user provide input? {is_input_provided}")
```
