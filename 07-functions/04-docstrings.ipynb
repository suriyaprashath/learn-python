## Docstring Syntax

- A docstring is a string literal (usually triple quotes) that appears as the first statement in a module, function, class, or method.
- It should describe what the object does.
- Multi-line docstrings can include details about parameters, return values, and examples.

**Basic syntax:**

```python
def function_name(params):
    """
    Brief description.
    
    Optional longer explanation.
    """
    # function body
```

## Real-World Example: Using Docstrings

Let's document a function that calculates the area of a circle.

````python
import math

def area_circle(radius):
    """
    Calculate the area of a circle.

    Parameters:
        radius (float): The radius of the circle.

    Returns:
        float: The area of the circle.
    """
    return math.pi * radius ** 2
````

````python
# Accessing the docstring
help(area_circle)
# area_circle.__doc__
````
