## Counting Occurrences

Dictionaries are often used to count how many times each item appears in a collection.

```python
text = 'hello world hello python world'
word_count = {}

for word in text.split():
    word_count[word] = word_count.get(word, 0) + 1

print(word_count)
```

## Grouping Data

Group items by a property using a dictionary of lists.

```python
students = [
    {'name': 'Alice', 'grade': 'A', 'subject': 'Math'},
    {'name': 'Bob', 'grade': 'B', 'subject': 'Math'},
    {'name': 'Charlie', 'grade': 'A', 'subject': 'Science'}
]
by_subject = {}

for student in students:
    subject = student['subject']
    by_subject.setdefault(subject, []).append(student['name'])
print(by_subject)
```

## Configuration and Nested Dictionaries

Dictionaries can store configuration data, including nested dictionaries.

```python
config = {
    'db': {'host': 'localhost', 'port': 5432},
    'debug': True
}

print(config['db']['host'])
```

## Lookup Tables

Use dictionaries to map keys to values for fast lookup.

```python
status = {200: 'OK', 404: 'Not Found'}

print(status.get(200))
print(status.get(500, 'Unknown'))
```
