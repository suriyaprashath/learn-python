# Escape Sequences in Python

Escape sequences are special characters used in strings to represent certain whitespace, symbols, or formatting that cannot be typed directly. They begin with a backslash (`\`). Here are some common escape sequences in Python:

- `\n` : New line  
    Example: `"Hello,\nWorld!"` prints:
    ```
    Hello,
    World!
    ```

- `\t` : Tab space  
    Example: `"Hello,\tWorld!"` prints:  
    `Hello,   World!`

- `\\` : Backslash  
    Example: `"Hello, \\World!"` prints:  
    `Hello, \World!`

- `\"` : Double quote  
    Example: `"Hello, \"World\"!"` prints:  
    `Hello, "World"!`

- `\'` : Single quote  
    Example: `'Hello, \'World\'!'` prints:  
    `Hello, 'World'!`

Escape sequences are useful for formatting strings and including special characters within them.

````python
# Examples of using escape sequences in print statements
print("Hello,\nWorld!")  # New line
print("Hello,\tWorld!")  # Tab space
print("Hello, \\World!")  # Backslash
print("Hello, \"World\"!")  # Double quotes
print('Hello, \'World\'!')  # Single quotes
````
