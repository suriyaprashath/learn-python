# Introduction to Modules in Python

Modules are files containing Python code (functions, classes, variables) that can be imported and reused in other programs. They help organize code, promote reuse, and make large projects easier to manage.

## Types of Modules

1. **Built-in Modules**: Provided by Python, such as `math`, `os`, and `sys`. No installation needed.
2. **Standard Library Modules**: Part of Pythonâ€™s standard library, offering a wide range of functionality (e.g., `datetime`, `random`, `json`).
3. **Third-party Modules**: Created by the community, installed via package managers like `pip` (e.g., `numpy`, `requests`).
4. **User-defined Modules**: Custom modules you create in your own files to organize your code.

## Examples of Different Types of Modules

```python
# Built-in module example: sys
import sys
print(sys.version)  # Shows Python version
```

```python
# Standard library module example: datetime
import datetime
print(datetime.datetime.now())  # Shows current date and time
```

```python
# Third-party module example: requests (must be installed with pip)
import requests

response = requests.get('https://api.github.com')
print(response.status_code)  # Shows HTTP status code
```

```python
# User-defined module example: my_module.py
import my_module
print(my_module.greet("Alice"))  # Hello, Alice!
```
