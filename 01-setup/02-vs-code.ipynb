# Install VS Code
Download URL - https://code.visualstudio.com/download

# VS Code Extensions for Python

Essential extensions to install:

1. **Python** (by Microsoft) - Python language support with IntelliSense, linting, debugging
1. **Jupyter** (by Microsoft) - Jupyter notebook support, interactive programming
1. **Pylance** (by Microsoft) - Fast, feature-rich language support for Python

## How to Install Extensions

1. Open Extensions view: `Ctrl + Shift + X`
1. Search for the extension name
1. Click "Install" button
1. Reload VS Code if prompted

# Useful Keyboard Shortcuts

- `Ctrl + Shift + P` - Command Palette
- `Ctrl + P` - Quick Open (find files)
- `Ctrl + B` - Toggle Sidebar
- ``Ctrl + ` `` - Toggle Terminal
- `Ctrl + /` - Toggle line comment
- `Alt + Up/Down` - Move line up/down
- `Shift + Alt + Up/Down` - Copy line up/down

# Settings and Customization

Access settings: `Ctrl + ,`

Some useful settings to configure:
- Font size: `editor.fontSize`
- Auto save: `files.autoSave` (set to `afterDelay`)
- Format on save: `editor.formatOnSave` (set to `true`)
- Python path: Configure in Settings or use Command Palette
