## The Problem: Unclear Code

Without type hints, it's hard to know what types a function expects or returns:

````python
# âŒ What types does this function expect?
def calculate_total(items, tax_rate):
    subtotal = sum(items)
    return subtotal * (1 + tax_rate)

# What should we pass in?
# Is items a list? A dict? Numbers or strings?
# Is tax_rate a percentage (8) or decimal (0.08)?
````

**Problems:**
- Unclear what data types are expected
- Easy to pass wrong types and get runtime errors
- No IDE autocomplete or helpful suggestions
- Code is harder to understand and maintain

---

## The Solution: Type Hints

Type hints (also called **type annotations**) let you specify what types your code expects. They make code more readable and help catch bugs early.

````python
# âœ… Clear what types are expected
def calculate_total(items: list[float], tax_rate: float) -> float:
    """Calculate total price with tax.
    
    Args:
        items: List of item prices
        tax_rate: Tax rate as decimal (e.g., 0.08 for 8%)
    
    Returns:
        Total price including tax
    """
    subtotal = sum(items)
    return subtotal * (1 + tax_rate)

# Now it's clear!
result = calculate_total([10.99, 5.50, 3.25], 0.08)
print(f"Total: ${result:.2f}")
````

**What happens:**
1. `:` after parameter name indicates the type (e.g., `items: list[float]`)
2. `->` before the colon indicates the return type
3. Your IDE can now show warnings if you pass wrong types
4. Other developers immediately understand what the function needs

## Basic Type Hints

### Simple Types

Python's basic types can be used directly:

````python
# Type hints for variables
name: str = "Alice"
age: int = 30
height: float = 5.6
is_student: bool = False

print(f"{name} is {age} years old")
````

````python
# Type hints for function parameters and return values
def greet(name: str, age: int) -> str:
    """Generate a greeting message."""
    return f"Hello, {name}! You are {age} years old."

message = greet("Bob", 25)
print(message)
````

**Key Points:**
- Type hints don't enforce types at runtime (Python is still dynamically typed)
- They're mainly for documentation and tooling (IDEs, linters)
- You can still pass wrong types, but your IDE will warn you

## Collection Types

For collections, specify both the container type and the element types:

````python
# List of strings
names: list[str] = ["Alice", "Bob", "Charlie"]

# Dictionary with string keys and int values
scores: dict[str, int] = {"Alice": 95, "Bob": 87, "Charlie": 92}

# Tuple with specific types (position matters)
person: tuple[str, int, float] = ("Alice", 30, 5.6)

# Set of integers
unique_ids: set[int] = {101, 102, 103}

print(f"Names: {names}")
print(f"Scores: {scores}")
print(f"Person: {person}")
````

### Example: Function with Collection Types

````python
def get_average_score(scores: dict[str, int]) -> float:
    """Calculate average score from a dictionary of scores.
    
    Args:
        scores: Dictionary mapping student names to their scores
    
    Returns:
        Average score as a float
    """
    if not scores:
        return 0.0
    return sum(scores.values()) / len(scores)

student_scores = {"Alice": 95, "Bob": 87, "Charlie": 92}
average = get_average_score(student_scores)
print(f"Average score: {average:.2f}")
````

## Practical Example: User Management System

````python
from typing import Optional

def create_user(
    username: str,
    email: str,
    age: int,
    bio: Optional[str] = None
) -> dict[str, str | int | None]:
    """Create a new user profile.
    
    Args:
        username: User's unique username
        email: User's email address
        age: User's age in years
        bio: Optional user biography
    
    Returns:
        Dictionary containing user profile data
    """
    return {
        "username": username,
        "email": email,
        "age": age,
        "bio": bio,
        "status": "active"
    }

# Create users
user1 = create_user("alice", "alice@example.com", 28)
user2 = create_user("bob", "bob@example.com", 32, "Python developer")

print(f"User 1: {user1}")
print(f"User 2: {user2}")
````

**What happens:**
1. Function clearly shows required parameters and their types
2. `bio` is optional (has default value `None`)
3. Return type shows dictionary structure
4. IDE provides autocomplete and type checking

## Best Practices

### âœ… Do:
- Use type hints for function parameters and return values
- Use type hints for complex variable assignments
- Use `Optional[Type]` or `Type | None` when values can be None
- Keep type hints simple and readable
- Use modern syntax (`|`) if using Python 3.10+

### âŒ Don't:
- Don't add type hints to every variable (obvious types don't need hints)
- Don't use type hints for very short, simple functions
- Don't use overly complex type hints that hurt readability
- Don't rely on type hints alone for validation (they're not enforced)

### Examples:

````python
# âœ… Good: Type hints improve clarity
def calculate_discount(price: float, discount_percent: float) -> float:
    """Calculate discounted price."""
    return price * (1 - discount_percent / 100)

# âœ… Good: Optional type hint for variable with complex type
user_permissions: dict[str, list[str]] = {
    "admin": ["read", "write", "delete"],
    "user": ["read"]
}

# âŒ Unnecessary: Type is obvious
count: int = 5  # Type is clear from the value

# âœ… Better: Skip hint for obvious types
count = 5
````

## Summary

### Key Concepts:
- **Type hints** make code more readable and help catch bugs
- Use `:` after parameters to specify types, `->` for return types
- Built-in types: `int`, `str`, `float`, `bool`
- Collection types: `list[T]`, `dict[K, V]`, `tuple[T, ...]`, `set[T]`
- `Optional[T]` or `T | None` for values that can be None
- `Union[A, B]` or `A | B` for multiple possible types
- Type hints are documentation, not runtime enforcement

### Syntax Reference:
```python
# Variable type hints
name: str = "Alice"
scores: list[int] = [95, 87, 92]

# Function type hints

    return f"Hello, {name}!"

# Optional and Union
def find(user_id: int) -> str | None:
    pass

def process(value: int | str) -> str:
    pass
```
