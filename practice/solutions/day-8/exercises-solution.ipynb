```python
"""Temperature conversion utilities."""

def celsius_to_fahrenheit(celsius: float) -> float:
    """Convert Celsius to Fahrenheit.
    
    Args:
        celsius: Temperature in Celsius
    
    Returns:
        Temperature in Fahrenheit
    """
    return (celsius * 9/5) + 32


def fahrenheit_to_celsius(fahrenheit: float) -> float:
    """Convert Fahrenheit to Celsius.
    
    Args:
        fahrenheit: Temperature in Fahrenheit
    
    Returns:
        Temperature in Celsius
    """
    return (fahrenheit - 32) * 5/9


def is_freezing(celsius: float) -> bool:
    """Check if temperature is at or below freezing point.
    
    Args:
        celsius: Temperature in Celsius
    
    Returns:
        True if temperature is at or below 0Â°C, False otherwise
    """
    return celsius <= 0
```

```python
"""Tests demonstrating fixtures and their importance."""
import pytest
from converter import celsius_to_fahrenheit, fahrenheit_to_celsius, is_freezing


# === Part 1: Define Fixtures ===

@pytest.fixture
def sample_celsius_temps():
    """Provide sample Celsius temperatures for testing.
    
    Returns:
        List of common Celsius temperatures
    """
    return [0, 25, 100, -10]


@pytest.fixture
def sample_fahrenheit_temps():
    """Provide sample Fahrenheit temperatures for testing.
    
    Returns:
        List of corresponding Fahrenheit temperatures
    """
    return [32, 77, 212, 14]


# === Part 2: Demonstrate Problem (Without Fixtures) ===

# âŒ Without fixtures (bad - repetitive setup)
def test_without_fixture_conversions():
    """Test conversions without fixtures - notice the repetition."""
    temps = [0, 25, 100, -10]  # Setup data manually
    
    results = [celsius_to_fahrenheit(t) for t in temps]
    assert len(results) == 4
    assert results[0] == 32  # 0Â°C = 32Â°F


def test_without_fixture_freezing():
    """Test freezing without fixtures - duplicating same data."""
    temps = [0, 25, 100, -10]  # Same data repeated!
    
    freezing_temps = [t for t in temps if is_freezing(t)]
    assert len(freezing_temps) == 2  # 0 and -10


# === Part 3: Solution (With Fixtures) ===

# âœ… With fixtures (good - reusable, clean)
def test_with_fixture_conversions(sample_celsius_temps):
    """Test conversions using fixture - no setup code!"""
    # Fixture automatically provides temps - no manual setup
    results = [celsius_to_fahrenheit(t) for t in sample_celsius_temps]
    assert len(results) == 4
    assert results[0] == 32


def test_with_fixture_freezing(sample_celsius_temps):
    """Test freezing using fixture - reusing same fixture."""
    # Same fixture, different test - DRY principle!
    freezing_temps = [t for t in sample_celsius_temps if is_freezing(t)]
    assert len(freezing_temps) == 2
    assert 0 in freezing_temps
    assert -10 in freezing_temps


def test_with_fixture_all_conversions(sample_celsius_temps, sample_fahrenheit_temps):
    """Test using multiple fixtures together."""
    # Can use multiple fixtures in one test!
    for celsius, expected_fahrenheit in zip(sample_celsius_temps, sample_fahrenheit_temps):
        result = celsius_to_fahrenheit(celsius)
        assert result == expected_fahrenheit
```

**Why Fixtures Are Important:**

**Without fixtures:**
- âŒ Duplicate setup code in every test
- âŒ Hard to maintain (change data in many places)
- âŒ Tests are longer and less focused

**With fixtures:**
- âœ… Setup code defined once at the top of the file
- âœ… Automatically injected into tests that request them
- âœ… Tests are clean and focused on assertions
- âœ… Easy to update data (change once in fixture)

**Key Points:**
- Fixtures are defined with `@pytest.fixture` decorator
- Simply add fixture name as test parameter
- pytest injects the fixture value automatically
- Multiple fixtures can be used in one test
- Each test gets a fresh copy of the fixture data


```python
"""Tests demonstrating pytest markers: skip and xfail."""
import sys
import pytest
from converter import celsius_to_fahrenheit, fahrenheit_to_celsius


# === 1. Unconditional Skip ===
@pytest.mark.skip(reason="Kelvin conversion not implemented yet")
def test_kelvin_conversion():
    """This test will always be skipped."""
    # Feature doesn't exist yet
    def celsius_to_kelvin(celsius):
        return celsius + 273.15
    
    result = celsius_to_kelvin(0)
    assert result == 273.15


# === 2. Conditional Skip - Python Version ===
@pytest.mark.skipif(sys.version_info < (3, 10), reason="Requires Python 3.10+ for union syntax")
def test_modern_union_syntax():
    """Only runs on Python 3.10+."""
    # Modern union syntax using |
    def get_temp(value: int | None) -> int | None:
        return value
    
    assert get_temp(25) == 25
    assert get_temp(None) is None


# === 3. Conditional Skip - Platform ===
@pytest.mark.skipif(sys.platform == "win32", reason="Not supported on Windows")
def test_unix_specific_feature():
    """Skipped on Windows, runs on Unix/Linux/Mac."""
    # Imagine this uses Unix-specific features
    result = celsius_to_fahrenheit(100)
    assert result == 212


# === 4. Expected Failure (XFail) ===
@pytest.mark.xfail(reason="Known bug with extreme temperatures")
def test_extreme_temperature():
    """This test is expected to fail due to a known limitation."""
    # Very large numbers might cause precision issues
    result = celsius_to_fahrenheit(1000000)
    # This assertion might fail, but that's expected
    assert result == 1800032.0  # Exact match might fail due to float precision


# === 5. XFail with Strict Mode ===
@pytest.mark.xfail(strict=True, reason="Must validate absolute zero")
def test_below_absolute_zero():
    """If this passes unexpectedly, the test fails (strict mode)."""
    # Should validate that temp is not below absolute zero (-273.15Â°C)
    # Currently we don't have this validation, so test might pass
    # But in strict mode, unexpected pass = failure
    
    # This should raise an error but doesn't (not implemented yet)
    result = celsius_to_fahrenheit(-300)  # Below absolute zero!
    assert result < -459.67  # -459.67Â°F is absolute zero


# === 6. XFail with Condition ===
@pytest.mark.xfail(sys.platform == "darwin", reason="Flaky on macOS")
def test_precision_sensitive():
    """Expected to fail on macOS due to floating point differences."""
    result = celsius_to_fahrenheit(37.7778)
    assert result == 100.0  # Might not match exactly


# === 7. Multiple Markers ===
@pytest.mark.skipif(sys.version_info < (3, 9), reason="Requires Python 3.9+")
@pytest.mark.xfail(reason="Known edge case")
```

**Understanding the Output:**

- **SKIPPED (s)**: Test was not executed

- **FAILED**: Test failed unexpectedly (with `strict=True` on xfail that passes)

**Why Use Markers:**

**Skip:**
- âœ… Feature not implemented yet
- âœ… Test only works on specific platforms
- âœ… Requires external resources (network, database)
- âœ… Temporarily disable flaky tests

**XFail:**
- âœ… Known bugs not yet fixed
- âœ… Document expected failures
- âœ… Don't break CI/CD pipeline
- âœ… With `strict=True`: Ensure bug gets fixed (fail if it passes)

## Exercise 6 Solution: Type Hinting - Basics, Limitations & Dataclasses

### Part 1: Why Type Hints Matter

````python
# Without type hints - unclear what types are expected
def calculate_total_no_hints(price, quantity):
    return price * quantity

# With type hints - clear and self-documenting
def calculate_total(price: float, quantity: int) -> float:
    """Calculate total price.
    
    Args:
        price: Unit price
        quantity: Number of items
    
    Returns:
        Total cost
    """
    return price * quantity

# Built-in types examples
def process_order(
    product_name: str,
    price: float,
    quantity: int,
    is_available: bool
) -> str:
    """Process an order with clear type hints."""
    if is_available:
        total = price * quantity
        return f"Order {product_name}: ${total:.2f}"
    return f"Sorry, {product_name} is not available"

# Test
result = process_order("Laptop", 999.99, 2, True)
print(result)
````

### Part 2: Advanced Types (Optional, List, Dict, Union)

````python
from typing import Optional

# Using Optional (value or None)
def greet_user(name: str, title: Optional[str] = None) -> str:
    """Greet user with optional title."""
    if title:
        return f"Hello, {title} {name}"
    return f"Hello, {name}"

# Using list type
def calculate_average(numbers: list[float]) -> float:
    """Calculate average of numbers."""
    return sum(numbers) / len(numbers)

# Using dict type
def get_total_price(items: dict[str, float]) -> float:
    """Sum all prices in dictionary."""
    return sum(items.values())

# Modern union syntax (Python 3.10+)
def process_items(
    items: list[str],
    quantities: dict[str, int],
    notes: str | None = None
) -> list[tuple[str, int]]:
    """Process items and return as tuples.
    
    Args:
        items: List of item names
        quantities: Dict mapping item names to quantities
        notes: Optional notes
    
    Returns:
        List of (item, quantity) tuples
    """
    result = [(item, quantities.get(item, 0)) for item in items]
    if notes:
        print(f"Note: {notes}")
    return result

# Test
items = ["apple", "banana", "orange"]
quantities = {"apple": 5, "banana": 3, "orange": 7}
result = process_items(items, quantities, "Fresh fruits")
print(result)
````

### Part 3: Type Hint Limitations

````python
# Type hints are NOT enforced at runtime!
def add_numbers(a: int, b: int) -> int:
    """Add two integers."""
    return a + b

# This should fail but doesn't!
result = add_numbers("hello", "world")
print(f"Result: {result}")  # "helloworld" - concatenation, not addition!
print(f"Type: {type(result)}")  # str, not int!

# Another example
def multiply(x: int, y: int) -> int:
    return x * y

# Accepts wrong types without error
print(multiply("abc", 3))  # "abcabcabc" - string repetition!
print(multiply(10.5, 2.5))  # 26.25 - float multiplication!

print("\nâš ï¸ Problem: Type hints are just documentation, not enforced!")
````

### Part 4: Dataclasses and Their Limitations

````python
from dataclasses import dataclass

@dataclass
class Product:
    """Product with type hints but no validation."""
    name: str
    price: float
    in_stock: bool

# Create with correct types
laptop = Product(name="Laptop", price=999.99, in_stock=True)
print(f"âœ… Valid product: {laptop}")

# Demonstrate limitation: accepts wrong types!
print("\nâš ï¸ Dataclass Limitations:")
invalid_product = Product(
    name=123,  # Should be str!
    price="not a number",  # Should be float!
    in_stock="maybe"  # Should be bool!
)
print(f"âŒ Invalid product created: {invalid_product}")
print(f"Name type: {type(invalid_product.name)}")
print(f"Price type: {type(invalid_product.price)}")
print("\nDataclasses don't validate types - they just document them!")
print("This is where Pydantic solves the problem...")
````

**Why This Solution Works:**
- âœ… **Type hints** improve code readability and IDE support
- âœ… **Advanced types** (Optional, List, Dict, Union) provide rich type information
- âŒ **Limitation**: Python doesn't enforce types at runtime
- âŒ **Dataclasses** provide structure but no validation
- âœ… **Next**: Pydantic solves these problems with automatic validation

## Exercise 7 Solution: Pydantic Basics - BaseModel & Type Validation

````python
from pydantic import BaseModel, ValidationError
from dataclasses import dataclass

# First, show the problem with dataclass (from Exercise 6)
@dataclass
class DataclassProduct:
    """Dataclass without validation."""
    name: str
    price: float

print("=== Dataclass Problem ===")
bad_product = DataclassProduct(name=123, price="not a number")
print(f"âŒ Dataclass accepts invalid data: {bad_product}")
print()

# Now, the Pydantic solution
class Product(BaseModel):
    """Pydantic model with automatic validation."""
    name: str
    price: float
    in_stock: bool

# 1. Create valid product
print("=== Pydantic Solution ===")
laptop = Product(
    name="Laptop",
    price=999.99,
    in_stock=True
)
print(f"âœ… Valid product created: {laptop}")
print(f"   Name: {laptop.name} (type: {type(laptop.name).__name__})")
print(f"   Price: {laptop.price} (type: {type(laptop.price).__name__})")
print(f"   In stock: {laptop.in_stock} (type: {type(laptop.in_stock).__name__})")
print()

# 2. Demonstrate automatic validation - wrong type for name
print("=== Testing Validation ===")
try:
    invalid_product1 = Product(
        name=123,  # Integer instead of string!
        price=999.99,
        in_stock=True
    )
except ValidationError as e:
    print("âŒ Validation Error caught (name as integer):")
    print(f"   Error: {e.errors()[0]['msg']}")
    print(f"   Field: {e.errors()[0]['loc']}")
print()

# 3. Try with wrong type for price
try:
    invalid_product2 = Product(
        name="Phone",
        price=[100, 200],  # List instead of float!
        in_stock=True
    )
except ValidationError as e:
    print("âŒ Validation Error caught (price as list):")
    print(f"   Error: {e.errors()[0]['msg']}")
    print(f"   Field: {e.errors()[0]['loc']}")
print()

# 4. Show all errors at once
try:
    invalid_product3 = Product(
        name=123,  # Wrong
        price="abc",  # Wrong
        in_stock="maybe"  # This might be coerced!
    )
except ValidationError as e:
    print("âŒ Multiple validation errors:")
    for error in e.errors():
        print(f"   Field '{error['loc'][0]}': {error['msg']}")
print()

# 5. Comparison summary
print("=== Comparison ===")
print("âœ… Pydantic: REJECTS invalid types")
print("âŒ Dataclass: ACCEPTS invalid types")
print("\nðŸŽ¯ Pydantic ensures data integrity at runtime!")
````

**Why This Works:**
- âœ… **BaseModel** provides automatic type validation
- âœ… **ValidationError** catches and reports type mismatches
- âœ… Validation happens **when creating instances** (runtime)
- âœ… Detailed error messages show what went wrong
- âœ… Solves the dataclass limitation problem

**Key Difference:**
- **Dataclass**: Type hints are documentation only
- **Pydantic**: Type hints are enforced rules

## Exercise 8 Solution: Type Coercion in Pydantic

````python
from pydantic import BaseModel, ValidationError

class Transaction(BaseModel):
    """Transaction model demonstrating type coercion."""
    transaction_id: int
    amount: float
    completed: bool
    description: str

# 1. Demonstrate successful coercion
print("=== Type Coercion Examples ===")

# String to int
trans1 = Transaction(
    transaction_id="12345",  # String "12345" â†’ int 12345
    amount=99.99,
    completed=True,
    description="Purchase"
)
print(f"âœ… String '12345' â†’ int {trans1.transaction_id}")
print(f"   Type: {type(trans1.transaction_id).__name__}")
print()

# String to float
trans2 = Transaction(
    transaction_id=67890,
    amount="149.99",  # String "149.99" â†’ float 149.99
    completed=False,
    description="Refund"
)
print(f"âœ… String '149.99' â†’ float {trans2.amount}")
print(f"   Type: {type(trans2.amount).__name__}")
print()

# Int to bool
trans3 = Transaction(
    transaction_id=11111,
    amount=50.00,
    completed=1,  # int 1 â†’ bool True
    description="Payment"
)
print(f"âœ… Integer 1 â†’ bool {trans3.completed}")
print(f"   Type: {type(trans3.completed).__name__}")
print()

trans4 = Transaction(
    transaction_id=22222,
    amount=75.50,
    completed=0,  # int 0 â†’ bool False
    description="Canceled"
)
print(f"âœ… Integer 0 â†’ bool {trans4.completed}")
print()

# Non-empty string to bool
trans5 = Transaction(
    transaction_id=33333,
    amount=100.00,
    completed="yes",  # Non-empty string â†’ True
    description="Approved"
)
print(f"âœ… String 'yes' â†’ bool {trans5.completed}")
print()

# Int to string
trans6 = Transaction(
    transaction_id=44444,
    amount=200.00,
    completed=True,
    description=789  # int 789 â†’ str "789"
)
print(f"âœ… Integer 789 â†’ string '{trans6.description}'")
print(f"   Type: {type(trans6.description).__name__}")
print()

# 2. Show what CANNOT be coerced
print("=== Failed Coercion Examples ===")

# String that cannot convert to int
try:
    trans_fail1 = Transaction(
        transaction_id="abc",  # Cannot convert to int
        amount=100.00,
        completed=True,
        description="Test"
    )
except ValidationError as e:
    print(f"âŒ Cannot coerce 'abc' to int")
    print(f"   Error: {e.errors()[0]['msg']}")
print()

# List cannot convert to float
try:
    trans_fail2 = Transaction(
        transaction_id=55555,
        amount=[10, 20],  # List cannot convert to float
        completed=True,
        description="Test"
    )
except ValidationError as e:
    print(f"âŒ Cannot coerce list [10, 20] to float")
    print(f"   Error: {e.errors()[0]['msg']}")
print()

# 3. Verify types after coercion
print("=== Type Verification ===")
test_trans = Transaction(
    transaction_id="99999",  # String
    amount="250.75",  # String
    completed=1,  # Integer
    description=555  # Integer
)
print(f"transaction_id: {test_trans.transaction_id} (type: {type(test_trans.transaction_id).__name__})")
print(f"amount: {test_trans.amount} (type: {type(test_trans.amount).__name__})")
print(f"completed: {test_trans.completed} (type: {type(test_trans.completed).__name__})")
print(f"description: {test_trans.description} (type: {type(test_trans.description).__name__})")
print()

print("ðŸŽ¯ Summary: Pydantic coerces compatible types, rejects incompatible ones!")
````

**Why This Works:**
- âœ… **Intelligent coercion**: Converts compatible types automatically
- âœ… String "123" â†’ int 123 (parseable)
- âœ… String "99.99" â†’ float 99.99 (parseable)
- âœ… Int 1/0 â†’ bool True/False
- âœ… Non-empty string â†’ bool True
- âŒ **Rejects invalid**: String "abc" cannot become int
- âŒ List cannot become float

**Key Insight:**
Pydantic makes data flexible (accepts common formats) while maintaining type safety (rejects nonsense).

## Exercise 9 Solution: Parsing/Loading from JSON and Dict

````python
from pydantic import BaseModel, ValidationError
import json

class User(BaseModel):
    """User model for parsing and validation."""
    user_id: int
    username: str
    email: str
    is_active: bool
    age: int | None = None  # Optional field

# 1. Parse from Dictionary
print("=== Parsing from Dictionary ===")
user_dict = {
    "user_id": 101,
    "username": "alice",
    "email": "alice@example.com",
    "is_active": True,
    "age": 28
}

user1 = User(**user_dict)  # Unpack dict and validate
print(f"âœ… User from dict: {user1}")
print(f"   Username: {user1.username}")
print(f"   Email: {user1.email}")
print()

# 2. Parse from JSON string
print("=== Parsing from JSON ===")
user_json = '''
{
    "user_id": 102,
    "username": "bob",
    "email": "bob@example.com",
    "is_active": false,
    "age": 35
}
'''

user2 = User.model_validate_json(user_json)
print(f"âœ… User from JSON: {user2}")
print(f"   User ID: {user2.user_id}")
print(f"   Active: {user2.is_active}")
print()

# 3. Demonstrate validation during parsing - invalid dict
print("=== Validation During Parsing ===")
invalid_dict = {
    "user_id": "not-a-number",  # Should be int
    "username": "charlie",
    "email": "charlie@example.com",
    "is_active": True
}

try:
    user3 = User(**invalid_dict)
except ValidationError as e:
    print("âŒ Dict validation failed:")
    print(f"   Error: {e.errors()[0]['msg']}")
    print(f"   Field: {e.errors()[0]['loc']}")
print()

# 4. Validate JSON with missing required field
invalid_json = '''
{
    "user_id": 103,
    "username": "diana"
}
'''

try:
    user4 = User.model_validate_json(invalid_json)
except ValidationError as e:
    print("âŒ JSON validation failed (missing fields):")
    for error in e.errors():
        print(f"   Missing field: {error['loc'][0]}")
print()

# 5. Export back to dict and JSON
print("=== Exporting Data ===")

# Export to dict
user_as_dict = user1.model_dump()
print(f"As dict: {user_as_dict}")
print(f"Type: {type(user_as_dict)}")
print()

# Export to JSON (compact)
user_json_compact = user1.model_dump_json()
print(f"As JSON (compact): {user_json_compact}")
print()

# Export to JSON (pretty)
user_json_pretty = user1.model_dump_json(indent=2)
print("As JSON (pretty):")
print(user_json_pretty)
print()

# 6. Demonstrate optional field
print("=== Optional Field Handling ===")
user_without_age = User(
    user_id=104,
    username="eve",
    email="eve@example.com",
    is_active=True
    # age is omitted - will be None
)
print(f"âœ… User without age: {user_without_age}")
print(f"   Age value: {user_without_age.age}")
print()

print("ðŸŽ¯ Pydantic validates data when parsing from dict/JSON!")
````

**Why This Works:**
- âœ… **Parse from dict**: `User(**dict)` - unpacks and validates
- âœ… **Parse from JSON**: `model_validate_json()` - parses and validates
- âœ… **Validation during parsing**: Catches errors when loading data
- âœ… **Export to dict**: `model_dump()` - converts back to dict
- âœ… **Export to JSON**: `model_dump_json()` - serializes to JSON string
- âœ… **Pretty printing**: `indent=2` for readable JSON

**Key Methods:**
| Operation | Method | Example |
|-----------|--------|--------|
| Dict â†’ Model | `**dict` | `User(**user_dict)` |
| JSON â†’ Model | `model_validate_json()` | `User.model_validate_json(json_str)` |
| Model â†’ Dict | `model_dump()` | `user.model_dump()` |
| Model â†’ JSON | `model_dump_json()` | `user.model_dump_json()` |

**Real-World Use Cases:**
- API responses (JSON â†’ Model)
- Database records (Dict â†’ Model)
- Config files (JSON/YAML â†’ Model)
- Data pipelines (Model â†’ Dict â†’ JSON)

## Exercise 10 Solution: Additional Validation - EmailStr & Custom Validators

````python
from pydantic import BaseModel, EmailStr, ValidationError, field_validator

# Part 1: EmailStr validation
class UserRegistration(BaseModel):
    """User registration with EmailStr and custom validators."""
    username: str
    email: EmailStr  # Built-in email validation
    age: int
    
    # Part 2: Custom validators
    @field_validator('username')
    @classmethod
    def validate_username(cls, value: str) -> str:
        """Validate and normalize username.
        
        Rules:
        - Minimum 3 characters
        - Maximum 20 characters
        - No spaces
        - Convert to lowercase
        """
        if len(value) < 3:
            raise ValueError("Username must be at least 3 characters")
        if len(value) > 20:
            raise ValueError("Username must be 20 characters or less")
        if ' ' in value:
            raise ValueError("Username cannot contain spaces")
        return value.lower()  # Transform to lowercase
    
    @field_validator('age')
    @classmethod
    def validate_age(cls, value: int) -> int:
        """Validate age range.
        
        Rules:
        - Minimum 13 years old
        - Maximum 120 years old
        - Must be positive
        """
        if value < 13:
            raise ValueError("User must be at least 13 years old")
        if value > 120:
            raise ValueError("Age must be 120 or less")
        if value < 0:
            raise ValueError("Age must be positive")
        return value


# === Test EmailStr Validation ===
print("=== EmailStr Validation ===")

# Valid email
try:
    user1 = UserRegistration(
        username="Alice",
        email="alice@example.com",
        age=25
    )
    print(f"âœ… Valid registration: {user1.username} ({user1.email})")
    print(f"   Note: Username converted to lowercase: '{user1.username}'")
except ValidationError as e:
    print(f"âŒ Error: {e}")
print()

# Invalid emails
invalid_emails = ["not-an-email", "user@", "@example.com"]
for invalid_email in invalid_emails:
    try:
        user = UserRegistration(
            username="test",
            email=invalid_email,
            age=25
        )
        print(f"âœ… Created with: {invalid_email}")
    except ValidationError as e:
        print(f"âŒ Invalid email '{invalid_email}': {e.errors()[0]['msg']}")
print()

# === Test Custom Validators ===
print("=== Custom Validator Tests ===")

# Test 1: Username too short
try:
    user2 = UserRegistration(
        username="ab",  # Only 2 characters
        email="user@example.com",
        age=20
    )
except ValidationError as e:
    print(f"âŒ Username too short: {e.errors()[0]['msg']}")
print()

# Test 2: Username with spaces
try:
    user3 = UserRegistration(
        username="john doe",  # Has space
        email="john@example.com",
        age=30
    )
except ValidationError as e:
    print(f"âŒ Username with spaces: {e.errors()[0]['msg']}")
print()

# Test 3: Age too young
try:
    user4 = UserRegistration(
        username="charlie",
        email="charlie@example.com",
        age=10  # Below minimum
    )
except ValidationError as e:
    print(f"âŒ Age too young: {e.errors()[0]['msg']}")
print()

# Test 4: Age too old
try:
    user5 = UserRegistration(
        username="diana",
        email="diana@example.com",
        age=150  # Above maximum
    )
except ValidationError as e:
    print(f"âŒ Age too old: {e.errors()[0]['msg']}")
print()

# Test 5: All validations pass
try:
    user6 = UserRegistration(
        username="ValidUser123",  # Will be lowercased
        email="valid@example.com",
        age=25
    )
    print("âœ… All validations passed!")
    print(f"   Username: {user6.username}")
    print(f"   Email: {user6.email}")
    print(f"   Age: {user6.age}")
except ValidationError as e:
    print(f"âŒ Error: {e}")
````

**Why This Works:**
- âœ… **EmailStr**: Built-in type that validates email format automatically
- âœ… **@field_validator**: Decorator for custom validation logic
- âœ… **@classmethod**: Required for field validators
- âœ… **ValueError**: Raised when validation fails
- âœ… **Value transformation**: Validators can modify values (e.g., lowercase)

**Custom Validator Pattern:**
```python
@field_validator('field_name')
@classmethod
def validate_field_name(cls, value: type) -> type:
```
