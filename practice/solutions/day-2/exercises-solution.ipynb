## Exercise 1: Age Category Checker - SOLUTION

```python
# Solution 1: Basic approach
age = 25

if age <= 12:
    print("Child")
elif age <= 19:
    print("Teenager")
elif age <= 59:
    print("Adult")
else:
    print("Senior")
```

```python
# Solution 2: More explicit ranges (easier for beginners to understand)
age = 25

if age >= 0 and age <= 12:
    print("Child")
elif age >= 13 and age <= 19:
    print("Teenager")
elif age >= 20 and age <= 59:
    print("Adult")
elif age >= 60:
    print("Senior")
```

```python
# Solution 3: With variable to store result (good practice)
age = 25
category = ""

if age <= 12:
    category = "Child"
elif age <= 19:
    category = "Teenager"
elif age <= 59:
    category = "Adult"
else:
    category = "Senior"

print(category)
```

## Exercise 2: Grade Calculator - SOLUTION

```python
# Solution 1: Straightforward approach
score = 85

if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
elif score >= 70:
    print("Grade: C")
elif score >= 60:
    print("Grade: D")
else:
    print("Grade: F")

if score >= 60:
    print("Passed!")
else:
    print("Failed!")
```

```python
# Solution 3: Using grade variable for pass/fail check
score = 85
grade = ""

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Grade: {grade}")

if grade != "F":
    print("Passed!")
else:
    print("Failed!")
```

## Exercise 4: Number Guessing Game - SOLUTION

```python
# Solution 1: Basic approach with break
target_number = 7
guess = 0

while guess != target_number:
    guess += 1
    print(f"Trying {guess}...")
    
    if guess == target_number:
        print("Found it!")
        break
    else:
        print("Not yet...")
```

```python
# Solution 2: Without break (simpler logic)
target_number = 7
guess = 0

while guess != target_number:
    guess += 1
    print(f"Trying {guess}...")
    
    if guess == target_number:
        print("Found it!")
    else:
        print("Not yet...")
```

```python
# Solution 3: With attempt counter
target_number = 7
guess = 0
attempts = 0

while guess != target_number:
    guess += 1
    attempts += 1
    print(f"Trying {guess}...")
    
    if guess == target_number:
        print(f"Found it in {attempts} attempts!")
    else:
        print("Not yet...")
```

## Exercise 5: Temperature Converter - SOLUTION

```python
# Solution 1: Basic function
def celsius_to_fahrenheit(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit

# Test
print(celsius_to_fahrenheit(0))    # 32.0
print(celsius_to_fahrenheit(100))  # 212.0
```

```python
# Solution 2: Direct return (more concise)
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# Test with formatted output
result = celsius_to_fahrenheit(0)
print(f"0°C is {result}°F")

result = celsius_to_fahrenheit(100)
print(f"100°C is {result}°F")
```

```python
# Solution 3: With multiple test cases
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# Test common temperatures
temperatures = [0, 25, 37, 100]

for temp in temperatures:
    result = celsius_to_fahrenheit(temp)
    print(f"{temp}°C = {result}°F")
```

## Exercise 6: Calculator Functions - SOLUTION

```python
# Solution 1: Basic implementation
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Cannot divide by zero"
    return a / b

# Test
print(add(10, 5))       # 15
print(subtract(10, 5))  # 5
print(multiply(10, 5))  # 50
print(divide(10, 5))    # 2.0
print(divide(10, 0))    # Cannot divide by zero
```

```python
# Solution 2: With better formatting
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Cannot divide by zero"
    return a / b

# Test with formatted output
print(f"10 + 5 = {add(10, 5)}")
print(f"10 - 5 = {subtract(10, 5)}")
print(f"10 × 5 = {multiply(10, 5)}")
print(f"10 ÷ 5 = {divide(10, 5)}")
print(f"10 ÷ 0 = {divide(10, 0)}")
```

```python
# Solution 3: With integer division option
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Cannot divide by zero"
    return a / b

def integer_divide(a, b):
    if b == 0:
        return "Cannot divide by zero"
    return a // b

# Test
print("Regular division:", divide(10, 3))      # 3.333...
print("Integer division:", integer_divide(10, 3))  # 3
```

## Bonus Exercise: Scope Explorer - SOLUTION

```python
# Solution 1: Basic global keyword usage
counter = 0

def increment_counter():
    global counter
    counter += 1
    print(f"Counter: {counter}")

# Test
increment_counter()  # Counter: 1
increment_counter()  # Counter: 2
increment_counter()  # Counter: 3
print(f"Final counter: {counter}")  # Final counter: 3
```

```python
# Solution 2: Demonstrating what happens WITHOUT global keyword
counter = 0

def increment_counter_wrong():
    # Without 'global', this creates a NEW local variable
    counter = 0  # Local variable (shadows global)
    counter += 1
    print(f"Inside function: {counter}")

increment_counter_wrong()  # Inside function: 1
print(f"Outside function: {counter}")  # Outside function: 0 (unchanged!)

print("\n--- Now with global keyword ---\n")

def increment_counter_correct():
    global counter
    counter += 1
    print(f"Inside function: {counter}")

increment_counter_correct()  # Inside function: 1
print(f"Outside function: {counter}")  # Outside function: 1 (changed!)
```

```python
# Solution 3: Better approach - return value instead of global
counter = 0

def increment_counter(current_value):
    """Better approach: take value as parameter, return new value"""
    new_value = current_value + 1
    print(f"Counter: {new_value}")
    return new_value

# Test - update counter with return value
counter = increment_counter(counter)  # Counter: 1
counter = increment_counter(counter)  # Counter: 2
counter = increment_counter(counter)  # Counter: 3
print(f"Final counter: {counter}")    # Final counter: 3

print("\nThis approach is preferred because:")
print("- No global keyword needed")
print("- Function is 'pure' (no side effects)")
print("- Easier to test and debug")
```

## Teaching Notes

### Common Student Mistakes:

1. **Conditionals:**
   - Forgetting the colon `:` after if/elif/else
   - Wrong indentation
   - Using `=` instead of `==` for comparison

2. **Loops:**
   - `range(10)` starts at 0, not 1 (common confusion)
   - Forgetting to increment in while loops (infinite loops)
   - Off-by-one errors with range

3. **Functions:**
   - Forgetting `return` statement
   - Calling function without parentheses
   - Not capturing return value in a variable

4. **Scope:**
   - Trying to modify global variable without `global` keyword
   - Confusion about when to use `global` vs return values

### Teaching Tips:

- **Exercise 1-2:** Show both approaches (explicit ranges vs chained comparisons)
- **Exercise 3:** Emphasize that `range(1, 11)` gives 1-10, not 1-11
- **Exercise 4:** Students often struggle with while loops - walk through the logic step by step
- **Exercise 5-6:** Stress the importance of `return` vs `print` in functions
- **Bonus:** Explain that `global` should be used sparingly; returning values is preferred
