**‚ö†Ô∏è Common mistakes:**

## Exercise 3: Coordinate System (Tuples)

**üéØ Learning objectives:**
- Create tuples
- Tuple unpacking
- Import and use modules

**‚ö†Ô∏è Common mistakes:**
- Forgetting to import math
- Wrong formula (forgetting parentheses in exponent)

```python
# Solution 1: Step-by-step
import math

# 1-2. Create tuples
point1 = (10, 20)
point2 = (30, 40)

# 3-4. Unpack tuples
x1, y1 = point1
x2, y2 = point2

# 5. Calculate distance
distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

# 6. Print results
print(f"Point 1: {point1}")
print(f"Point 2: {point2}")
print(f"Distance: {distance:.2f}")
```

```python
# Solution 2: More compact
import math

point1 = (10, 20)
point2 = (30, 40)
x1, y1 = point1
x2, y2 = point2

distance = math.sqrt((x2-x1)**2 + (y2-y1)**2)

print(f"Point 1: {point1}")
print(f"Point 2: {point2}")
print(f"Distance: {distance:.2f}")
```

```python
# Solution 3: Advanced (direct unpacking in formula)
import math

point1 = (10, 20)
point2 = (30, 40)

# Can unpack directly in expressions (advanced!)
distance = math.sqrt((point2[0]-point1[0])**2 + (point2[1]-point1[1])**2)

print(f"Point 1: {point1}")
print(f"Point 2: {point2}")
print(f"Distance: {distance:.2f}")
```

## Exercise 4: Remove Duplicates (Sets)

**üéØ Learning objectives:**
- Convert list to set and back
- Set operations (union, intersection, difference)
- Understand when to use sets

**‚ö†Ô∏è Common mistakes:**
- Forgetting sorted() when converting back to list
- Confusing set operators (& vs | vs -)
- Not understanding sets are unordered

```python
# Solution 1: Step-by-step with explanations

# 1. Original list with duplicates
numbers = [1, 2, 3, 2, 4, 5, 1, 6, 3, 7, 4]
print(f"Original list: {numbers}")

# 2. Convert to set (removes duplicates)
unique_numbers = set(numbers)

# 3. Convert back to sorted list
sorted_numbers = sorted(unique_numbers)
print(f"Unique sorted: {sorted_numbers}")

# 4. Create two sets
set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}

# 5. Set operations
# Intersection - items in BOTH sets
common = set_a & set_b
print(f"\nCommon elements: {common}")

# Union - ALL unique items from both sets
all_unique = set_a | set_b
print(f"All unique elements: {all_unique}")

# Difference - items in set_a but NOT in set_b
only_a = set_a - set_b
print(f"Only in set_a: {only_a}")
```

```python
# Solution 2: Using method syntax (alternative)

numbers = [1, 2, 3, 2, 4, 5, 1, 6, 3, 7, 4]
print(f"Original list: {numbers}")
print(f"Unique sorted: {sorted(set(numbers))}")

set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}

print(f"\nCommon elements: {set_a.intersection(set_b)}")
print(f"All unique elements: {set_a.union(set_b)}")
print(f"Only in set_a: {set_a.difference(set_b)}")
```

## Exercise 5: Grade Book (Lists + Dictionaries)

**üéØ Learning objectives:**
- Work with list of dictionaries
- Calculate aggregate statistics
- Find maximum value

**‚ö†Ô∏è Common mistakes:**
- Not initializing variables before loop
- Division by zero when calculating average
- Comparing wrong values when finding max
- Logic errors in counting conditions

```python
# Solution 1: Step-by-step

# 1. Create list of student dictionaries
students = [
    {"name": "Alice", "score": 85},
    {"name": "Bob", "score": 92},
    {"name": "Charlie", "score": 78},
    {"name": "Diana", "score": 95}
]

# 2. Print each student
print("Student Grades:")
for student in students:
    print(f"{student['name']}: {student['score']}")

# 3. Calculate average
total_score = 0
for student in students:
    total_score += student['score']

average = total_score / len(students)
print(f"\nAverage score: {average}")

# 4. Find highest score
highest_score = 0
highest_student = ""

for student in students:
    if student['score'] > highest_score:
        highest_score = student['score']
        highest_student = student['name']

print(f"Highest score: {highest_student} with {highest_score}")

# 5. Count students with 80+
count_80_plus = 0
for student in students:
    if student['score'] >= 80:
        count_80_plus += 1

print(f"Students with 80+: {count_80_plus}")
```

```python
# Solution 2: Using built-in functions (more Pythonic)

students = [
    {"name": "Alice", "score": 85},
    {"name": "Bob", "score": 92},
    {"name": "Charlie", "score": 78},
    {"name": "Diana", "score": 95}
]

print("Student Grades:")
for student in students:
    print(f"{student['name']}: {student['score']}")

# Using sum() for average
scores = [s['score'] for s in students]
average = sum(scores) / len(scores)
print(f"\nAverage score: {average}")

# Using max() with key parameter
top_student = max(students, key=lambda s: s['score'])
print(f"Highest score: {top_student['name']} with {top_student['score']}")

# Using sum() with condition (generator expression)
count_80_plus = sum(1 for s in students if s['score'] >= 80)
print(f"Students with 80+: {count_80_plus}")
```

```python
# Solution 3: Single loop (most efficient)

students = [
    {"name": "Alice", "score": 85},
    {"name": "Bob", "score": 92},
    {"name": "Charlie", "score": 78},
    {"name": "Diana", "score": 95}
]

# Do everything in one loop (efficient!)
print("Student Grades:")
total = 0
highest_score = 0
highest_name = ""
count_80_plus = 0

for student in students:
    name = student['name']
    score = student['score']
    
    print(f"{name}: {score}")
    total += score
    
    if score > highest_score:
        highest_score = score
        highest_name = name
    
    if score >= 80:
        count_80_plus += 1

print(f"\nAverage score: {total / len(students)}")
print(f"Highest score: {highest_name} with {highest_score}")
print(f"Students with 80+: {count_80_plus}")
```

## Exercise 6: Word Counter (Dictionaries + Loops)

**üéØ Learning objectives:**
- Use dictionary for counting
- Master the get() pattern for counting
- Find max value in dictionary

**‚ö†Ô∏è Common mistakes:**
- Not understanding the get(key, default) pattern
- Trying to access key before it exists
- Wrong logic for finding max

```python
# Solution 1: Using get() pattern

sentence = "the quick brown fox jumps over the lazy dog the fox"

# 1-2. Split and create counter dictionary
words = sentence.split()
word_count = {}

# 3. Count words using get() pattern
for word in words:
    # get() returns 0 if word not in dict, otherwise returns current count
    word_count[word] = word_count.get(word, 0) + 1

# 4. Print counts
print("Word counts:")
for word, count in word_count.items():
    print(f"{word}: {count}")

# 5. Find most common word
max_count = 0
max_word = ""

for word, count in word_count.items():
    if count > max_count:
        max_count = count
        max_word = word

print(f"\nMost common word: '{max_word}' appears {max_count} times")
```

```python
# Solution 2: Using if-else (alternative approach)

sentence = "the quick brown fox jumps over the lazy dog the fox"
words = sentence.split()
word_count = {}

# Check if word exists before counting
for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

print("Word counts:")
for word, count in word_count.items():
    print(f"{word}: {count}")

# Using max() with key parameter (advanced)
most_common = max(word_count, key=word_count.get)
print(f"\nMost common word: '{most_common}' appears {word_count[most_common]} times")
```

```python
# Solution 3: Using Counter from collections (advanced)
from collections import Counter

sentence = "the quick brown fox jumps over the lazy dog the fox"
words = sentence.split()

# Counter does all the counting automatically!
word_count = Counter(words)

print("Word counts:")
for word, count in word_count.items():
    print(f"{word}: {count}")

# most_common() method gives top N items
most_common_word, count = word_count.most_common(1)[0]
print(f"\nMost common word: '{most_common_word}' appears {count} times")
```

## Exercise 7: Random Data Generator (Modules)

**üéØ Learning objectives:**
- Import and use modules
- Use random module functions
- Use math module functions

**‚ö†Ô∏è Common mistakes:**
- Forgetting to import modules
- Wrong function names (randint vs random)
- Not understanding shuffle modifies in-place

```python
# Solution 1: Step-by-step
import random
import math

# 2. Random number between 1-100
random_num = random.randint(1, 100)
print(f"Random number: {random_num}")

# 3. List of 5 random numbers 1-50
random_list = []
for i in range(5):
    random_list.append(random.randint(1, 50))
print(f"Random list: {random_list}")

# 4. Random choice from list
fruits = ["apple", "banana", "cherry", "date"]
random_fruit = random.choice(fruits)
print(f"Random fruit: {random_fruit}")

# 5. Shuffle a list
numbers = list(range(1, 11))  # Creates [1, 2, 3, ..., 10]
random.shuffle(numbers)  # Modifies list in-place!
print(f"Shuffled: {numbers}")

# 6. Math calculations
sqrt_result = math.sqrt(144)
power_result = math.pow(2, 8)
print(f"\nSquare root of 144: {sqrt_result}")
print(f"2^8: {power_result}")
```

```python
# Solution 2: More compact (list comprehension)
import random
import math

print(f"Random number: {random.randint(1, 100)}")

# List comprehension to generate random list
random_list = [random.randint(1, 50) for _ in range(5)]
print(f"Random list: {random_list}")

fruits = ["apple", "banana", "cherry", "date"]
print(f"Random fruit: {random.choice(fruits)}")

numbers = list(range(1, 11))
random.shuffle(numbers)
print(f"Shuffled: {numbers}")

print(f"\nSquare root of 144: {math.sqrt(144)}")
print(f"2^8: {math.pow(2, 8)}")
```

## üéØ Bonus Exercise: Contact Book (All Concepts)

**üéØ Learning objectives:**
- Combine all Day 3 concepts
- Work with nested data structures
- Create reusable functions

```python
# Solution 1: Complete implementation

# 1. Create contacts list
contacts = [
    {"name": "Alice", "phone": "555-1234", "email": "alice@email.com", "tags": {"friend", "work"}},
    {"name": "Bob", "phone": "555-5678", "email": "bob@email.com", "tags": {"family"}},
    {"name": "Charlie", "phone": "555-9012", "email": "charlie@email.com", "tags": {"friend", "school"}}
]

# 2. Print all contacts formatted
print("Contact List:")
print("="*13)
for contact in contacts:
    print(f"Name: {contact['name']}")
    print(f"Phone: {contact['phone']}")
    print(f"Email: {contact['email']}")
    print(f"Tags: {', '.join(sorted(contact['tags']))}")
    print()

# 3. Find all unique tags (using set union)
all_tags = set()
for contact in contacts:
    all_tags = all_tags | contact['tags']  # Union

print(f"All tags: {all_tags}")

# 4. Find contacts with "friend" tag
friends = []
for contact in contacts:
    if "friend" in contact['tags']:
        friends.append(contact['name'])

print(f"\nFriends: {', '.join(friends)}")

# 5. Create find_contact function
def find_contact(name):
    """Find contact by name and return (phone, email) tuple."""
    for contact in contacts:
        if contact['name'] == name:
            return (contact['phone'], contact['email'])
    return (None, None)

# Test the function
phone, email = find_contact("Bob")
print(f"\nLooking up Bob: ({phone}, {email})")
```

```python
# Solution 2: More Pythonic version

contacts = [
    {"name": "Alice", "phone": "555-1234", "email": "alice@email.com", "tags": {"friend", "work"}},
    {"name": "Bob", "phone": "555-5678", "email": "bob@email.com", "tags": {"family"}},
    {"name": "Charlie", "phone": "555-9012", "email": "charlie@email.com", "tags": {"friend", "school"}}
]

# Print contacts
print("Contact List:")
print("="*13)
for c in contacts:
    print(f"Name: {c['name']}\nPhone: {c['phone']}\nEmail: {c['email']}")
    print(f"Tags: {', '.join(sorted(c['tags']))}\n")

# All unique tags using set.union()
all_tags = set().union(*[c['tags'] for c in contacts])
print(f"All tags: {all_tags}")

# Friends using list comprehension
friends = [c['name'] for c in contacts if "friend" in c['tags']]
print(f"\nFriends: {', '.join(friends)}")

# Find contact function with next()
def find_contact(name):
    contact = next((c for c in contacts if c['name'] == name), None)
    if contact:
        return (contact['phone'], contact['email'])
    return (None, None)

print(f"\nLooking up Bob: {find_contact('Bob')}")
```
