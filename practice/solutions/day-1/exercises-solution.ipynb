## Exercise 1: String Formatting - SOLUTION

```python
# Solution
title = "Python Crash Course"
author = "Eric Matthes"
year = 2019
price = 29.99

# Method 1: Multiple print statements
print(title.upper())
print(f"by {author} ({year})")
print(f"Price: ${price:.2f}")

print()  # Blank line

# Method 2: Single formatted string
output = f"""{title.upper()}
by {author} ({year})
Price: ${price:.2f}"""
print(output)
```

## Exercise 2: User Input Calculator - SOLUTION

```python
# Solution
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

print("\nResults:")
print(f"{num1} + {num2} = {num1 + num2}")
print(f"{num1} - {num2} = {num1 - num2}")
print(f"{num1} * {num2} = {num1 * num2}")
print(f"{num1} / {num2} = {num1 / num2:.2f}")
print(f"{num1} // {num2} = {num1 // num2}")
print(f"{num1} % {num2} = {num1 % num2}")

# Alternative: Store results in variables
addition = num1 + num2
subtraction = num1 - num2
multiplication = num1 * num2
division = num1 / num2
floor_division = num1 // num2
modulus = num1 % num2

print("\nAlternative format:")
print(f"Addition: {addition}")
print(f"Subtraction: {subtraction}")
print(f"Multiplication: {multiplication}")
print(f"Division: {division:.2f}")
print(f"Floor Division: {floor_division}")
print(f"Modulus: {modulus}")
```

## Exercise 3: String Manipulation - SOLUTION

```python
# Solution
sentence = "Python programming is fun and powerful"

print(f"Original: {sentence}")
print(f"Uppercase: {sentence.upper()}")
print(f"Lowercase: {sentence.lower()}")
print(f"Title Case: {sentence.title()}")

# Word count
words = sentence.split()
word_count = len(words)
print(f"Word Count: {word_count} words")

# Replace word
new_sentence = sentence.replace("programming", "coding")
print(f"After Replace: {new_sentence}")

# Check if word exists
contains_fun = "fun" in sentence
print(f"Contains 'fun': {contains_fun}")

# Additional operations
print(f"\nBonus operations:")
print(f"Starts with 'Python': {sentence.startswith('Python')}")
print(f"Ends with 'powerful': {sentence.endswith('powerful')}")
print(f"Find 'fun': {sentence.find('fun')}")
print(f"Count 'a': {sentence.count('a')}")
```

## Bonus Challenge - Contact Card - SOLUTION

```python
# Solution
name = input("Enter name: ").strip().title()
email = input("Enter email: ").strip().lower()
phone = input("Enter phone: ").strip()

# Display formatted card
separator = "═" * 40
print(separator)
print("CONTACT CARD".center(40))
print(separator)
print(f"Name:  {name}")
print(f"Email: {email}")
print(f"Phone: {phone}")
print(separator)

# Alternative with better formatting
print("\nAlternative format:")
card = f"""
{'═' * 40}
{'CONTACT CARD':^40}
{'═' * 40}
{'Name:':<10} {name}
{'Email:':<10} {email}
{'Phone:':<10} {phone}
{'═' * 40}
"""
print(card)
```

## Key Concepts Demonstrated

### String Methods Used:
- `.upper()`, `.lower()`, `.title()` - Change case
- `.strip()` - Remove whitespace
- `.split()` - Split string into list
- `.replace()` - Replace substring
- `.find()`, `.count()` - Search operations
- `.startswith()`, `.endswith()` - Check prefix/suffix

### F-String Features:
- Basic: `f"{variable}"`
- Expressions: `f"{variable.upper()}"`
- Formatting: `f"{price:.2f}"` (2 decimal places)
- Alignment: `f"{text:^40}"` (center in 40 chars)
- Alignment: `f"{text:<10}"` (left align in 10 chars)

### Type Conversion:
- `int()` - Convert to integer
- `float()` - Convert to float
- `str()` - Convert to string

### Operators:
- Arithmetic: `+`, `-`, `*`, `/`, `//`, `%`, `**`
- Comparison: `==`, `!=`, `<`, `>`, `<=`, `>=`
- Membership: `in`, `not in`
- String: `*` (repetition), `+` (concatenation)
