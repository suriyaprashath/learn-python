   - Attribute: `charge_level` (starts at 100)
   - Method: `charge()` - sets charge to 100
   - Method: `use(hours)` - reduces charge by `hours * 10`

2. **Phone class:**
   - Attribute: `battery` (a Battery object - composition)
   - Method: `check_battery()` - prints current charge level
   - Method: `use_phone(hours)` - uses the battery for given hours

**Example:**
```python

phone.use_phone(3)
```

### Exercise 10: Count Calls Decorator

**Task:** Create a decorator that counts how many times a function is called.

**Requirements:**
- Create a `count_calls` decorator
- Track the number of times the decorated function is called
- Print "Function {name} has been called {count} times" after each call
- Apply it to a function and test it

**Example:**
```python
@count_calls
def greet(name):
    print(f"Hello, {name}!")

greet("Alice")  # Function greet has been called 1 times
greet("Bob")    # Function greet has been called 2 times
```
