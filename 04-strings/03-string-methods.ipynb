# Introduction to String Methods in Python

String methods are built-in functions that allow you to perform various operations on strings, such as changing case, searching, replacing, splitting, and more. These methods make it easy to manipulate and analyze text data efficiently. In Python, string methods are called using the dot (.) notation after a string object.

Below are examples of some commonly used string methods.

## Changing Case

```python
# Changing case
text = "Python Programming"

print(text.upper())        # 'PYTHON PROGRAMMING'
print(text.lower())        # 'python programming'
print(text.title())        # 'Python Programming'
print(text.capitalize())   # 'Python programming'
print(text.swapcase())     # 'pYTHON pROGRAMMING'
```

## Searching and Replacing

```python
# Searching and counting
text = "Python Programming"

print(text.find("Pro"))           # 7 (index of substring)
print(text.count("m"))            # 2 (number of times 'm' appears)
print(text.startswith("Python"))  # True
print(text.endswith("ing"))       # True
```

```python
# Replacing text
text = "Python Programming"

print(text.replace("Python", "Java"))  # 'Java Programming'
```

## Splitting and Joining

```python
# Splitting and joining
text = "Python Programming Language"

words = text.split()           # ['Python', 'Programming', 'Language']
joined = "-".join(words)       # 'Python-Programming-Language'

print(words)
print(joined)

# Splitting with a delimiter
csv_data = "apple,banana,cherry"
fruits = csv_data.split(",")   # ['apple', 'banana', 'cherry']
print(fruits)
```

## Removing Whitespace

```python
# Stripping whitespace
text = "   Python Programming   "

print(text.strip())     # 'Python Programming'
print(text.lstrip())    # 'Python Programming   '
print(text.rstrip())    # '   Python Programming'
```

## Type Checking Methods

```python
# String type-checking methods
text1 = "Python"
text2 = "12345"
text3 = "Python3"
text4 = "   "

print(text1.isalpha())   # True (all letters)
print(text2.isdigit())   # True (all digits)
print(text3.isalnum())   # True (letters and numbers)
print(text4.isspace())   # True (all whitespace)
```
